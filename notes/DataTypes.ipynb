***Built-in Data Types***

- Variables can store data of different types which do different things

- Python (like all languages) have the following data types built-in by default

    1. Text Type:	str
    2. Numeric Types:	int, float, complex
    3. Sequence Types:	list, tuple, range
    4. Mapping Type:	dict
    5. Set Types:	set, frozenset
    6. Boolean Type:	bool
    7. Binary Types:	bytes, bytearray, memoryview
    8. None Type:	NoneType

- the type() function helps identify the data type

- data type is set to a variable at assignment


```python
#getting data type
alpha = 3.45994
print(type(alpha))

#data type is set at assignment
alpha_list = ['beta', 'gamma', 'delta']
alpha_tuple = ('theta','eta','kappa')

#or you can specify data types this way
espilson = float(5)

#other type examples
dictionary = {"name" : "John", "age" : 36}

#explore frozen sets
```
