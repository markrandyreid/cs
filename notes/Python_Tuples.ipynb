# Python Tuples

### Basic characteristics of Tuples: 
- used to storage multiple items (any data type) in a single variabe
- indexed, (defined) ordered and unchangeable with permission for duplicates.
- can contain a single item so long as a comma is at the end
- Syntax 
```new_tuple = ('This','is','an','example')```
or using the tuple constructor
```new_tuple = tuple(('This','is','an','example))```


`````python
#create tuple
new_tuple = ('This','is','an','example')

#print lenght of tuple
print(len(new_tuple))

#print tuple
print(new_tuple)

#print array type
print(type(new_tuple))
`````

### Accessing Tuple Items
- since tuples are indexed, access is simliar to lists
- negative indexes and ranges apply to tuples
- Use ````in```` keyword to check if an item exists in a tuple. Example below:

`````python
#tuple access via index and negative index examples
#new tuple
tuple_index = ('this', 'is','an','example','of','accessing','items','in','a','tuple','using','indexes')

#access and print item 3
print(tuple_index[2])

#access and print the last 6 items
print(tuple_index[-5:])

#access and print items 3 through 6
print(tuple_index[2:6])

#check if item is in tuple using in keyword
if 'using' in tuple_index:
    print('Yes, the word using is located in this tuple, but the index position is currently unknown')
`````

### Updating Tuples
- Can't change tuples
- workaround: change tuple to a list, invoke change, then turn list back into a tuple
- as a list, you can use various list methods to make the changes necessary
- Note: you can add a tuple to an existing tuple

`````python
#creating a random tuple
example_tuple1 = ("apple", "banana", "cherry",'pear')

#converting tuple into a list
example_tuple1_2_list = list(example_tuple1)

#adding new element to the list
example_tuple1_2_list.append('orange')

#converting back to a tuple 
example_tuple1 = tuple(example_tuple1_2_list)
print(type(example_tuple1))
print(example_tuple1)
`````

### Unpark Tuples
- unpacking essentially means assigning a variable to an item in a tuple
- or extracting an item to a variable
- note: # of variables = # of items in tuple, if not an * must be used to collect the remaining values as a list.



`````python
#creating a random tuple
example_tuple2 = ("apple", "mango", "papaya", "pineapple", "cherry")

#assign variables to reach item in the tuple and use the asterisks
#in this example, we'll assign items 1-3 and 5

(green, orange, *remove_soon,red) = example_tuple2

#print variables to check assignments
print(green,orange,remove_soon,red)

#note again that use of the arterisk gathers relevant items into a list by that title.
`````

### Looping thru Tuples
- same as list: we can loop thru each item or loop by index position
- see examples: 

`````python
example_tuple3 = ("apple", "mango", "papaya", "pineapple", "cherry",'banana')

#by item
for x in example_tuple3:
    print(x)

#by position or index
for i in range(len(example_tuple3)):
    print(example_tuple3[i]*2)

#while loop
i = 0

#while i is greater than the number of items in this tuple
while i < len(example_tuple3):
    #print each item and a new line
    print(f'{example_tuple3[i]}\n')
    #after each iteration add 1
    i += 1
`````

### Joining Tuples
- use the addition operator to join two or more tuples

`````python
#example adding tuples
example_tuple4 = example_tuple1 + example_tuple2
print(example_tuple4)
`````

### Multiply Tuples
- use the multiplication operator

`````python
#example multipling tuples
example_tuple5 = 10*example_tuple1
print(len(example_tuple1))
print(len(example_tuple5))
`````

### Tuple Methods
- two methods for tuples in python
    - ````count()```` - returns the number of times a specific value occurs in a tuple
    - ````index()```` - searches tuple for specific value and returns value position
        - if specific value occurs more than once in the tuple, this method will return the position of the first occurance. 
    


`````python
#example of count() method
example_tuple5.count('apple')
`````

`````python
#example of index() method
example_tuple4.index('pear')
`````
