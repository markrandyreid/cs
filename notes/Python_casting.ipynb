**Casting** - using a specific DATA TYPE on to a variable

Constructor functions used to cast: 
- int(): constructs integer number from integer, float or string literal (assuming the string is a whole number)
- float(): constructs float number from integer, float or string literal (assuming the string is a whole number)
- str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals

```python
#example

alpha = int(5.99) #builds integer number from float literal
#built rounds inputs down

beta = float(3.4) #builds float number from float literal

x = str(304.30) #builds string from float literal

print(f"{alpha}\n{beta}\n{type(x)}")
```
