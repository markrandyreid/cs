#   Fizz Buzz (Loop and Condition Exercise)
- Create a program that will run through the integers 1 to 25:
    - If the number is a multiple of 3, print "Fizz"
    - If the number is a multiple of 5, print "Buzz"
    - If the number is a multiple of 3 and 5 print "FizzBuzz"
    - If the number is not any of the stated multiples, print the number.
    - Each output is on it's own line

- Also include in your repo a .drawio file that flowcharts out the logic

```python
for x in range(1,26):
    if x % 3 == 0 and x % 5 == 0:
        print('FizzBuzz')
    #use ELIF to evaluate a new mutually execlusive condition against x when the previous condition is false
    elif x % 3 == 0: 
        print('Fizz')
    elif x % 5 == 0: 
        print('Buzz')
    #use ELSe here to catch the remaining values of x that fail previous conditions
    else:
        print(x)
```
