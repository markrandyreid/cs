**Python Conditions and If Statements**
- Based on logical conditions from mathematics
- structure: evaluate condition, if true, execute code block inside the if statement, if false SKIP the code block (or execute the ELSE statement)

```python
#example
alpha = 12
beta = 14

if alpha >= beta:
    print(f'alpha greater than or beta')
else: 
    print(f'beta > alpha')
```

**Indentation**
- used to define scope in the code
- same amount of indentation for all statements within the same code block

```python
#example: 
age = int(input('How old are you:'))

if age >= 18:
    print(f'You are an adult')
    print(f'who can vote')
```

**The Elif Keyword**
- python's way of saying "if the previous CONDITIONS were note true, then try this condition"
- allows coder to check if multiple expressions are True, then execute code block as soon as one of the conditions is True
- when Elif is used, python evaluates the conditions from top to bottom. Once it finds a condition that is true, that code block is executed and it skips evaluating the remaining conditions
- impact on memory usage

***When to use Elif***
- when you have multiple conditions that are mutually exclusive

```python
#example
def pickaday(day):
    if day == 1:
        print("Monday")
    elif day == 2:
        print("Tuesday")
    elif day == 3:
        print("Wednesday")
    elif day == 4:
        print("Thursday")
    elif day == 5:
        print("Friday")
    elif day == 6:
        print("Saturday")
    elif day == 7:
        print("Sunday")

pickaday(5)
```

**Python Else Statement**
- ELSE keyword catches anything that isn't caught by preceding conditions (previous if or elif are false)
- ELSE keyword can be used without elif creating a simple two way choice
- Known as the default "catch all" action when non of the previous conditions are true
    - as such the else statement must come last.



```python
#example
'''
Validating user input:
'''

username = input('Please provide your last name: ')

if len(username) > 0:
  print(f"Welcome, {username}!")
else:
  print("Error: Username cannot be empty")
```

**Shorthand if Statements**
- Only use when: 
    - The condition and actions are simple
    - It improves code readability
    - You want to make a quick assignment based on a condition
    - Different syntax compared to the normal usages: 
        - variable = value_if_true if condition else value_if_false
        - colon is still needed after the condition

```python
#example: 
alpha = 3094
beta = 309493
gamma = alpha if alpha < beta else beta
print(f'The larger variable has a value of: {gamma}')
```

```python
#example
'''Multiple Conditions on One Line'''
x = 15
y = 20
max_value = x if x > y else y
print("Maximum value:", max_value)

a = 2
b = 5
print('Yes') if a == b else print('No')
```
