# <span style="color:red">**Conditional Logic Assignment**</span>

Create a program that will ask the user for a number between 1 and 12. You may assume the input is correct.

After getting the number display which season it is. The ranges are:
- Spring: 3,4,5
- Summer: 6,7,8
- Fall: 9,10,11
- Winter 12,1,2

Here is a sample output:
What month is it? (1-12): 2
"The current season is Winter."

***For an extra 2 points, display the month as all. So the output becomes:***
What month is it? (1-12): 2
"The month is February and the current season is Winter."

**Remember to also complete the flowchart. It is strongly advised that you do the flowchart first, as this will help you write the code.**

# Code Walkthru Answer with Notes

```python
#request user intput
x = int(input('What month is it? (1-12): '))

#create dictionary assignning keys to numbers 1 - 12 and values to months
number_month = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

#conditions based on question parameters

if x >= 1 and x <= 2 or x == 12:
    print('The month is', number_month[x], 'and the current season is Winter.')

#use elif because we only need to run one of these conditions
elif x >= 3 and x <= 5:
    print('The month is', number_month[x], 'and the current season is Spring.')
elif x >= 6 and x <= 8:
    print('The month is', number_month[x], 'and the current season is Summer.')
elif x >= 9 and x <= 11:
    print('The month is', number_month[x], 'and the current season is Fall.')
```
