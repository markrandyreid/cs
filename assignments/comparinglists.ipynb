# Comparing Lists
### Problem Instructions: 
- Create two separate lists for player one and player two, each containing 10 random numbers between 1 and 50. Do not sort the lists. Compare the lists in order, with whichever list having the higher number winning that round. Keep track of how many times each list wins. 
- Find the highest number in each list and its index, taking the first instance if the number occurs multiple times. 
- Find the lowest number in each list and its index, also taking the first instance if the number occurs multiple times. 
    - A tie is not recorded as a win for either player. 
- Display the lists, then report to the user how many times each player won and the information about the highest and lowest numbers with their indices. 

### Output Example: 
````
Player One = [5,7,2,9,1,1,3,8,6,9]
Player Two = [3,8,5,5,8,1,4,7,4,7]
Player one won 5 times
Player two won 4 times
Player one's highest number is 9 at index 3
Player two's highest number is 8 at index 1
Player one's lowest number is 1 at index 4
Player two's lowest number is 1 at index 5
````

### Answer (Version 1)

`````python
import random

p1_list = []
p2_list = []
p1_wins = 0
p2_wins = 0
ties = 0

for x in range(0,10):
    p1_list.append(random.randint(1,50))  
    p2_list.append(random.randint(1,50))

#for each round of comparsion
for y in range(len(p1_list)):
  if p1_list[y] > p2_list[y]:
    p1_wins += 1
  if p1_list[y] < p2_list[y]:
    p2_wins += 1
  else:
    ties += 1

print(f'Player 1 List: {p1_list}')
print(f'Player 2 List: {p2_list}')
print(f'Player 1 Wins: {p1_wins} times')
print(f'Player 2 Wins: {p2_wins} times')
print(f'Player 1 Highest #: {max(p1_list)} (Index {p1_list.index(max(p1_list))})')
print(f'Player 2 Highest #: {max(p2_list)} (Index {p2_list.index(max(p2_list))})')
print(f'Player 1 Lowest #: {min(p1_list)} (Index {p1_list.index(min(p1_list))})')
print(f'Player 2 Lowest #: {min(p2_list)} (Index {p2_list.index(min(p2_list))})')
`````
