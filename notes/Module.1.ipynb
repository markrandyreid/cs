## 1. Definition of Artificial Intelligence (AI)

### Key Definitions:
- **Max Tegmark**: AI is “intelligence that is not biological.”
- **John McCarthy**: AI is the science and engineering of making intelligent machines.
- **Nils J. Nilsson**: AI involves making machines intelligent—able to function appropriately and with foresight in their environment.
- **Rodney Brooks**: AI is not a single entity but a collection of practices and components.

### Summary:
- AI mimics human intelligence through software.
- It's a broad field that involves problem-solving, adaptability, and the creation of intelligent behavior in machines.
- AI includes various tools and methods assembled to make machines “think” like humans.


## 2. Prominent AI Scientists

### John McCarthy:
- Coined the term "Artificial Intelligence" (1955).
- Developed **LISP**, a key AI programming language.
- Founded Stanford AI Lab; won Turing Award (1971).

### Alan Turing:
- Father of modern computer science.
- Created the **Turing Machine** and **Turing Test**.
- Played a crucial role in WWII by decoding Enigma.

### Turing Test:
- Evaluates whether a machine can mimic human conversation indistinguishably.
- ChatGPT has reportedly passed versions of this test.


## 3. AI vs Automation

| Feature | Automation | Artificial Intelligence |
|--------|------------|--------------------------|
| Focus | Routine tasks | Complex, human-like cognition |
| Learning | No learning | Learns and adapts |
| Programming | Explicit rules | Learns from data (training) |
| Human Role | Minimal | Requires supervision/training |
| Job Impact | Replaces jobs | Transforms or creates jobs |

**Example**: Rule-based systems vs. trained AI that recognizes images or speech.


## 4. ANI vs AGI

### Artificial Narrow Intelligence (ANI):
- Specialized in one task (e.g., ChatGPT, Alexa).
- Examples:
  - **Siri, Alexa** – voice assistance
  - **ChatGPT** – conversation
  - **DALL-E** – image generation
  - **AlphaGo** – board games
  - **Gemini** – multi-modal AI (text, images)

### Artificial General Intelligence (AGI):
- Human-level intelligence across many tasks.
- Self-aware, capable of general problem solving (e.g., identify, drive, compose, converse).

### Artificial Superintelligence (ASI):
- Exceeds human abilities in all areas (theoretical).

#### Judging AI Systems:
- **Performance** (compared to a human expert).
- **Generality** (number of tasks performed).


## 5. Limitations of AI

1. **Data Dependency**:
   - Needs large, clean, unbiased datasets.
   - Biases in training data → biased outputs.

2. **Cost**:
   - Training LLMs is expensive (e.g., A100 GPUs at $20k each).
   - Example: BloombergGPT trained over 1.3M hours.

3. **Transparency & Explainability**:
   - AI decision-making is often opaque.
   - Critical in sectors like finance and healthcare.

4. **Ethical & Safety Issues**:
   - Privacy violations, biased decisions, surveillance risks.
   - Safety concerns in transport, military, etc.

5. **Creativity & Emotion**:
   - Can mimic but not truly possess them.

6. **Common Sense & Consciousness**:
   - AI lacks true understanding of context and common sense.


## 6. History of AI: Timeline

### 1950s–1980s: Foundational Period
- **1950**: Turing's paper on machine intelligence.
- **1956**: Dartmouth Conference (birth of AI).
- **1960s–1980s**: Expert systems (ELIZA, Prolog, CNNs); first “AI Winter” (decline in funding).

### 1990s–2000s: Resurgence & Big Data
- **1997**: Deep Blue beats Kasparov.
- **2006**: Hinton revives deep learning.
- **2011–2012**: Watson wins Jeopardy; Google’s self-driving car logs 300K miles.

### 2010s–2020s: Deep Learning & Ethics
- **2012**: AlexNet wins ImageNet.
- **2014**: GANs by Ian Goodfellow.
- **2016**: AlphaGo defeats Lee Sedol.
- **2019**: GPT-2 released.
- **2020–2024**: GPT-3, ChatGPT, AlphaFold, Gemini released; AI ethics becomes a focus.


## 7. Evolution of AI

### Neural Networks (1950–70s):
- Inspired by the human brain.
- Perceptron by McCulloch & Pitts.

### Machine Learning (1980–2010):
- Learns patterns from data to predict outcomes.
- Arthur Samuel coined the term.

### Deep Learning (2010–present):
- Trains massive neural networks on big data.
- Powers current AI tools like GPT and image recognition systems.
