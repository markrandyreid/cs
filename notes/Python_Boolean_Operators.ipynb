***Assigment Operators***

The Walrus Operator (:=)
- assign a value to a variable as part of an expression (computes and stores value)
- previously you could only assign values in statements, not inside expressions. 

```python
#example

# Without walrus
my_list = {1,2,3,4,5,6,23}

value = len(my_list)
if value > 5:
    print(f"List has {value} elements")

# With walrus
if (value := len(my_list)) > 5:
    print(f"List has {value} elements")
```

***Identity Operators***
- used to compare if objects are the same with the same memory location
- is AND is not
- this checks identity not value (==)

**Difference Between is and ==**
 - is - Checks if both variables point to the same object in memory
 - == - Checks if the values of both variables are equal

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print(x == y)
```

***Membership Operators**
- in : returns true if a sequence with the specified value is present in the object
- not in : : returns true if a sequence with the specified value is not present in the object
- membership operators are case-sensitive

```python
#example
my_list = {12,13,4,5,65,12,34,15,95,23}

#string example: 
sample_text = 'mark randy reid'

text = "Hello World"

print("H" in text)
print("hello" in text)
print("z" not in text)
print(65 in my_list)
print(3049 not in my_list)
print('x' in sample_text)
```
