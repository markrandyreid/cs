**For Loops**
- primary function is to iterate over a sequence presented as a list, tuple, dictionary, string or set
- technically kwown as an iteration method simliar to thoses in other object-oreintated programming languages.

***Looping thru a string***
- strings are iterable ojects as they contain a sequence of characters
- as such, we can use a for loop to manipulate the string

```python
#example

name = 'Mark Randy Reid'
#for every character in the name variable, print each character on a seperate line
for i in name: 
    print(f'{i}')
```

**The Break Statement**
- BREAK statements stop any loop before it has looped through all item
- Not that the break stops the loop where the break condition is true but still executes that code so anything preceding that point will not execute. 

```python
#example

names = ['mark','randy','reid', 'the','second','estate','of','Wellington']
for x in names:
    print(x)
    if x == 'the':
        break
```

**The continue Statement**
- used to stop the current iteration of the loop, but continue the next one.
- a condition skip to the next one

```python
#example
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

**The range() function**
- function returns a sequence of numbers starting at 0 (by default), increasing by 1 and ends at a specific number (input - 1)
- syntax: range(start, stop, increments of)

```python
#example
#list of multiples of 2 until 20
for x in range(2,21,2):
    print(x)
```

**The ELSE in For Loop**
- the ELSE keyword in a for loop specifies a block of code to be executed when the loop is fisihed. 
- Note - the else block will NOT executed if the loop is stopped by a break statement.

```python
#example
'''Else statement executed'''
for x in range(5):
  if x == 3: 
    print(x)
else:
  print("Finally finished!")

'''
Else statement not executed.
'''
for x in range(6):
  if x == 3: break
  print(x)
else:
  print("Finally finished!")
```

**Nested Loops**
- The typical structure involves an outer loop controlling the overall iterations and an inner loop executing fully during each cycle of the outer loop. This leads to nested iterations, where the total number of executions is a product of the iteration counts of both loops.
- allows for interation over complex, multidimensional data structures such mas matrices or grids. 
- Performance considerations: 
    - increases execution time (Execution Time), especially if both loops have large iteration cycles
        - Note the differences between Run and Exection Time: 
            - Execution time specifically measures the actual amount of time it takes for a program or a particular code segment to complete its execution. 
            - Runtime refers to the entire phase during which a program is actively running. This includes not only the actual execution of the program's instructions but also program initialization, memory allocation, input/output operations, waiting for user inputs, and cleanup tasks after execution. Runtime is a broader concept encompassing the whole lifecycle from the moment the program starts executing until it terminates, including environment and system interactions.
    - efficient use and optimization are key to avoiding bottlenecks.
- Execution Flow: Understanding the execution flow of nested loops is crucial: the inner loop executes fully before the outer loop moves to its next iteration. This means the total iterations equal the product of iterations of all nested levels. 

```python
#example: 
data = ['alpha','beta','gamma','delta','sigma']

for x in range(10):              #outloop for rows
    for i in range(7):           #innerloop for columns executes completely for every outloop iteration
        print('*', end=' ')
    print()                      #prints to next line after each outloop iteration
```
