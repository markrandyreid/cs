```python
'''
Find the cost of the meal for a customer based on their age and drink order. 
'''

print('Welcome to the Menu Order System')

customerAge = int(input('What is the age of the customer? '))

if customerAge <= 12:
    price = 8
#use elif because first condition is not met
elif customerAge >= 62:
    price = 12  
#use else as a catch all for any other conditions
else:
    price = 12
print('The cost of this customer is $', price)

drinkYesNo = input('Would you like a drink? (Y/N) ')

if drinkYesNo == 'Y':
    small_Large = input('Small or Large? (S/L) ')
    if small_Large == 'L':
        price += 2
    else:
        price += 1

print('Total Cost $', price)
    
```
