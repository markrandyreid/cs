## What is a list?

- items can be any data type
- lists are one of four built-in data types used to store collections of data: 
    - other three are tuples, sets and dictionaries

`````python
#creates and prints list
#note two different ways to create lists
alpha = [304,234,2304,2030,'akkre','beta']

#usign the list() constructor 
beta = ('mark','randy','reid')
list(beta)

print(alpha)
print(beta)
`````

## List Items
- Items within a list are ordered, changeable (mutable) and can be duplicate values 
    - CONSIDER: performance ramifications on lists vs. tuples?
- lists items are indexed:
    - ```new_list[0]```
- New items are added to the end of a list unless list methods are used to change list items (see below) 
- DATA TYPE: List items can be any data type: string, int, boolean, float, etc

`````python
#examples of accessing list items by index

#print item based on index location
alpha = [304,234,2304,2030,'akkre','beta']

#print first item 
print(alpha[0])

#print last item 
print(alpha[-1])

#add item to the list, then print index to check addition
alpha.append(2405)
print(alpha[-1])
`````

## Python Collections

## Python - Access List Itemsm
- items are indexed meaning they can be accessed by referring to an index number
- Negative indexing
    - starts from the end 
    - example_list[-1] pulls the last item
    - exmaple_list[-2] pulls the second to last item

### Range of Indexes
- you can specific a range of indexes by specifying where to start and end the range
- the return valuye will be a new list with the specified items
- Note: search will begin at the start value and end (but not include) the last value.

`````python
#example of range of indexes
original_list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(original_list[0:3])

#returning from index to end
print(original_list[3:])

#returning values from the end to a specific index in the list
print(original_list[-4:-1])
`````

## Changing List Items
- changing the value of an item requires referring to it's index number or range
- Note: this method eliminates the existing item and replaces it with a new item


`````python
#changing list items
numbers = []
for i in range(1,15,2):
    numbers.append(i)

print(numbers)
#numbers[1:5]

numbers[0] = 'Odd Numbers Excluding 1'
print(numbers)

#changing range of items
range_numbers = []
for x in range(1,100,10):
    range_numbers.append(x)
print(range_numbers)

range_numbers[0:6] = [0,0,0,0,0,0]
print(range_numbers)
`````

##  Methods to ADD items to lists

### insert() Method 
- insert() method inserts an item at a specific index
- Note: this method adds item to a specific index location, increaing the length of list and displacing the existing item to the left. 
- since this is a method, we add to the list_variable with a period
    - Syntax: new_list.insert(index location, new item)

### append() method
- method adds items to the end of a list 
    - Syntax: new_list.append(new item)

### extend() method
- used to merge lists or any iterable object such as tuples, sets, dictionaries and sets
- after extend is used, what data type is the new object?
    - Syntax: new_list.extend(new iterable)




`````python
example_list = []
example_list2 = []

#example for appending items to end of a list
for x in range(10,101,10):
    example_list.append(x)
print(example_list)

#insert method example
example_list.insert(0,20)
print(example_list)

#extend method example
for i in range(100,201,10):
    example_list2.append(i)
print(example_list2)

example_list.extend(example_list2)
print(example_list)
`````

## Removing List Items
### remove() method
- method removes a specific item based on it's syntax and not location
    - syntax: new_list.remove('delete')

### pop() and del() method
- pop method removes the specified index (and associated item)
    - syntax: new_list.pop(34)
- del method also does the same thing
    - syntax: new_list.del[34]
    - Note del() method can also be used to del the entire list
        - del new_list

### clear() method
- method removes items within the list and results in an empty list
    - syntax: new_list.clear(34)





## Loop Lists
- for loops can be used to loop thru a list
- you can also loop thru a list using the index number using len() and range() functions

`````python
example_list3 = []
#example of loop lists
for i in range(1,100,10):
    example_list3.append(i*2)
print(example_list3)

#examle of looping thru list with index
for x in range(len(example_list3)):
    print(example_list3[x]*2)
`````

### Using a While Loop
- using the len() function to determine the length of the list, start at 0 then loop through the list items by referring to their indexes. 
- remember to increase by one after every interation

`````python
#example of using while loop and list
#list to loop through
example_list3

i = 0
while i < len(example_list3):
    print(example_list3[i]*2)
    i = i + 1
`````

### List Comprehension

### List comprehension to manipulate strings before new list creation
- the expression is the current item in iteration but it is also the ourcome, which can be manipulated beofr it ends up in the new lists

`````python
#example
#manipulate existing values
new_fruits = [x.upper() for x in fruits]
print(new_fruits)

#changing values in the list
new_fruits2 = ['Bad Food' for x in new_fruits]
print(new_fruits2)

#changing values based on condition in new list
new_fruits3 = [x if x!= 'banana' else 'orange' for x in fruits]
print(new_fruits3)
`````

### Sorting Lists
- sort method is used to reorganize list items alphanumerically, ascending, descending,etc. 
- this method is case sensative so the uppercase strings will be sorted before lower case strings. 
    - to neutralize this, you can use the builtin method ```object.lower()``` to perform a case insensative sort
- Ascending, by default
    - Syntax: ```object.sort()```
    - or ```object.sort(reverse=True)```
- this method can be customized using the keyword argument, where you can set a function as the keyword argument and sort by the rules of that function
    - syntax: ```object.sort(key=some_function)```
- to reverse the list, use the reverse method
    - Syntax: ```object.reverse()```

`````python
#examples

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

#sort
fruits.sort()
print(fruits)

#sort decsending
fruits.sort(reverse=True)
print(fruits)

#sort by function
def myfunct('x'):
    return x.upper
`````
