# Python Sets

### Introduction
- single variable used to store multiple items
    - Data type: string, boolean, int, float
    - sets may contain different data types
- sets are immutable, unindexed and unordered (can't be accessed by index or key)
- no duplicate values 
- new items can be added once a set is created
- Syntax: ````example_set = {'for', 'the', 'live'}````
- Syntax: ````exmaple_set = set(('for', 'the', 'live'))````


`````python
#example of sets
example_set0 = {32, 24, 204.304, 'wealth','power'}
example_set1 = set((32.1, 24.2, 204.302, 'wealthy','powerful'))
print(type(example_set0))
print(len(example_set0))
print(type(example_set1))
print(len(example_set1))
`````

### Joining Sets (Review)


`````python

#example of differences
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)

print(set3)

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.difference_update(set2)

print(set1)
`````

### Frozen Sets
- truly immutable version of a set
- items are unique, unordered and unchangable
- any iterable can be converted to a frozen set
- syntax: ````random_set = frozenset(example_set1)````

`````python
#example
example_frozenset = frozenset(example_set0)
print(example_frozenset)
print(type(example_frozenset))

#access
for x in example_frozenset:
    print(x)
`````

### Frozen Set Methods
- ````copy()````: Returns a shallow copy
- ````difference()````: Returns a new frozenset with the difference	between two sets
- ````intersection()```` or ````&````: Returns a new frozenset with the intersection	
- ````isdisjoint()````: Returns whether two frozensets have an intersection
- ````issubset()```` or ````<= / <````: Returns True if this frozenset is a (proper) subset of another
- ````issuperset()```` or ````>= / >````: Returns True if this frozenset is a (proper) superset of another	
- ````symmetric_difference()```` or ````^````: Returns a new frozenset with the symmetric differences
- ````union()```` or ````|````: Returns a new frozenset containing the union

`````python
#examples
a = example_frozenset.copy()
print(a)
`````

#### Set Methods
- ````add()````: Adds an element to the set
- ````clear()````: Removes all the elements from the set
- ````copy()````: Returns a copy of the set
- ````difference()````: Returns a set containing the difference between two or more sets
- ````difference_update()````: Removes the items in this set that are also included in another
- ````discard()````: Removes the specific item  
- ````intersection()```` (&): Returns a set, that is the intersection of two other sets (duplicate values)
- ````intersection_update````(&=): Removes the items in this set that are not present in other, specified set(s)
- ````isdisjoint````: returns booolean whether two sets have an intersection or not
- ````issubset()````(<=): Returns True if all items of this set is present in another set or 
(<) Returns True if all items of this set is present in another, larger set

- ````issuperset()```` (>:): Returns True if all items of another set is present in this set or (>) Returns True if all items of another, smaller set is present in this set
- ````pop()````:Removes an element from the set
- ````remove()````: Removes the specified element
- ````union()````(|): Return a set containing the union of sets
- ````update()````(|): Update the set with the union of this set and others
- ````symmetric_difference()````(^): Returns a set with the symmetric differences of two sets or items that are not present in both sets
- ````symmetric_difference_update()````(^=): Inserts the symmetric differences from this set and another or removes items present in both sets, and inserts the items not present in both sets. 

`````python
#Symmetric Difference examples
x = {"apple", "banana", "cherry",}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)
zeta = x.symmetric_difference_update(y)
print(z)
print(zeta)

`````

`````python
#symmetric difference update example 
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x ^= y
print(x)
`````

`````python
#difference example
alpha = []
beta=[]
for x in range(0,51,5):
    alpha.append(x)

for x in range(0,51,10):
    beta.append(x)

set_alpha = set((alpha))
set_beta = set((beta))

print(set_alpha)
print(set_beta)

#difference
ab_diff = set_alpha.difference(set_beta)
print(ab_diff)
`````
