**Python Numbers**
- int
- floats
- complex (written with a j as the imaginary part)

```python
x = 3+5j
y = 5j
z = -5j

print(type(x))
print(type(y))
print(type(z))
```

**Conversion**
- int to float adds .0
- float to int omits decimal value
- int or float adds imaginary number + 0j


```python
x = 1    # int
y = 2.8  # float
z = 1j   # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

#convert from float to complex:
d = complex(y)

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
```
