# Dictionary

### Introduction
- used to storage data values in key:value pairs.
- ordered (a specific defined order) and changable, but no duplicates.
- values can be referred to using the key name
- syntax: ````example_dict = {'Name': Mark, 'Age': 35, 'Location': 'United Kindom'}````
- constructir syntax: ````thisdict = dict(name = "John", age = 36, country = "Norway")````
- data type: dictionary items can be any data type.
- len(dict) prints the number of pairs in dictionary


`````python
#example
example_dict0 = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]}

print(example_dict0)
print(len(example_dict0))
`````

### Accessing Dictionary Items
- we can use the following methods to access items in a dictionary: 
    - Get method: ````example_dict.get('key_value')````
    - Direct: ````example_dicti['key_value]````

- We can also get the keys and values
    - ````example_dict.values()````: returns the values of a dictionary as a list
    - ````example_dict.keys()````: returns the keys of a dictionary as a list
    - ````example_dict.items()````: returns (key:values) or items of dictionary as a tuple

- We can check if specific keys within a dictionary exists using the ````in```` keyword

`````python
#example of accessing dictionary items
print(example_dict0.get('brand'))
print(example_dict0['year'])
print(example_dict0.values())
print(example_dict0.keys())
print(example_dict0.items())
`````

### Changing Dictionary Items
- change the value of a specific item by using it's key name
    - ````example_dict0['brand'] = 'Tesla'````
- the update method could accomplish the same change
- Note: if the item doesn't exist, it will be added to the dictionary
    - ````example_dict0.update({'brand': 'Tesla'})````

`````python
#example 
example_dict0.update({'brand': 'TESLA'})
print(example_dict0.values())
`````

### Add Dictionary Items
- We can use the index key and assign the value
    - ````example_dict0["color"] = "red"````
- We can also use the  update method since it will add to the dictionary if the item doesn't exist
    - ````example_dict0.update({'make': 'model y'})````

`````python
#example
example_dict0.update({'make': 'model y'})
example_dict0["color"] = "red"
print(example_dict0.items())
print(len(example_dict0))
`````

### Remove Dictionary Items
- Several methods to remove items fromd dictionary
    - ````pop()```` removes item wit a specified key name
        - ````del```` doesnt the same thing and can also delete the dictionary 
    - ````popitem()```` removes the last inserted item
    - ````clear()```` method emptoes the dictionary 

`````python
#example
example_dict0.pop('make')
print(example_dict0)

example_dict0.popitem()
print(example_dict0)

`````

### Looping thru Dictionaries
- return values can be items, keys or values spending on how the loop is structure

`````python
#example of looping dictionary for keys only
for x in example_dict0:
    print(f'{x}\n')

#example of looping dictionary for values
for x in example_dict0.values():
    print(x)

#example of looping dictionary for items
for x,y in example_dict0.items():
    print(x,y)
    
`````

### Copying Dictionaries
- use the build in method ````example_dict0.copy()````

`````python
#example
example_dict1 = example_dict0.copy()
print(example_dict1)
`````

### Nested Dictionaries
- simply put, it is a dictionary within a dictionary.

### Accessing Items in a nested dictionary
- start with the outer most dictionary and work your way to the item
    - ````outer_dict[first_dict][key]

`````python
#example of nested dictionary and access

myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

#access child 2, year
print(myfamily['child2']['year'])
`````

`````python
### Looping through Nested Dictionaries
# this is tuple unpacking
# list all birth years


for child_id, child_info in myfamily.items():
    print(f' Year of Birth: {child_info['year']}\n')

# filtering while looping
family = {
    'child1': {'name': 'Emily', 'age': 5},
    'child2': {'name': 'John', 'age': 8},
    'child3': {'name': 'Sarah', 'age': 12},
    'child4': {'name': 'Mike', 'age': 15}
}

#outer dictionary, inner dictionary in out dictionary items
for child, info in family.items():
    if info['age'] <= 12:
        print(f'{info['name']} is {info['age']}')


#modifications via loop
#changing ages
for child, info in family.items():
    info['age']*=2

print(family.items())

#creating a list from data in dictionary
names = []

for child, info in family.items():
    names.append(info['name'])

print(names)
`````
