## Mutability

* Immutable types: `str`, `tuple` (as well as all other we learned!)
* Mutable types: `list`

**Objects of mutable types can be modified once they are created.**

```python
lst1=[1,2,3,'2',8,6]
print(lst1)
lst1[2]=100
print (lst1)
```

```python
tp1=(1,2,3,'2',7,9)
print(tp1)
tp1[2]=100
print (tp1)
```

*   Lists take more momory - Not an issue!!
*   Lists sometimes have lower performance - Not an issue!!

### Mutability Can Be Quite Convenient

Popular methods:

* `L.append(e)`
* `L.extend(L1)`

There are several other useful list methods, see http://docs.python.org/3/library/stdtypes.html#mutable-sequence-types:


### List Methods: `.append` vs. `.extend`

```python
list_alpha = [1, 2, 3]
list_alpha.append(4) #adds another number to the end of the list
print(list_alpha)

#extend
list_alpha.extend([5, 6, 7, 8, 9, 10]) #merging or concatenation
print(list_alpha)
```

```python
#appending a list to a list
#append only adds one element, therefore to add a list to a list, you should use extend
list_beta = [1, 2, 3]
list_beta.extend([4,5,6])
print(list_beta)
```

#### Example using `.append`

Form a list which contains numbers from 1 to 100

```python
i=0                   #starting point
lst1=[]               #empty list to store values
while(i<100):          #until this condition is false
  #print(i,end='')
  lst1.append(i)      #do this, appending the new value of i after every iteration=
  i=i+1

print(lst1)
```

Form a list which contains numbers from 1 to 100 that are divisible by 5

```python
i=5
lst1=[]
while(i<101):
  #print(i,end='')
  lst1.append(i)
  i=i+5
print(lst1)
```

### Mutability Can Also Be Dangerous

```python
L1 = [1, 2, 3]
L2 = [4, 5, 6, 7]

print('L1: ', L1)
print('L2: ', L2)

L1.append(L2[:]) #clone it #previous method appends as one item / one list this makes L1 connects to L2, so when L2 changes L1 will also change (aliases)
print('L1: ', L1)
print('L2: ',L2)

L2.append(100) #this sees L2 as two parts [1,2,3] and [4,5,6,7]
print('L1: ', L1)
print('L2: ',L2)
```

####Aliasing vs Cloning

```python
L1 = [1, 2, 3]
L2 = L1
L3 = L1[:] # Using slicing[:] is one way to clone the value without the permanent connection

print("L1 =", L1,"L2 =",L2,"L3 =",L3)

L1.append(4)
print("L1 =", L1,"L2 =",L2,"L3 =",L3)
```
