**While Loop**
- executes a set of statements as long as the WHILE condition is true
- requires relevant variables to be ready
- BREAK statements can stop a while loop even if the condition still holds true
- CONTINUE statements just stop the current iteration of the loop and moves on to the next
- ELSE is used to execute code once the while statement is no longer true


```python
#example

average = 87

while average < 100: 
    print(average)
    average += 1
else:
    print('Score is too high')
```

**Lecture Notes**
- While statements are ideal for situations where the number of iterations is unknown.
